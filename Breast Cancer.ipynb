```python
X,Y=data_frame.loc[:,:'worst fractal dimension'],data_frame['label']
```

```python
X.shape
```

```python
X.drop('mean perimeter',axis=1,inplace=True)
X.drop('mean area',axis=1,inplace=True)
X.drop('mean concave points',axis=1,inplace=True)
X.drop('mean symmetry',axis=1,inplace=True)
X.drop('mean fractal dimension',axis=1,inplace=True)
X.drop('radius error',axis=1,inplace=True)
X.drop('texture error',axis=1,inplace=True)
X.drop('perimeter error',axis=1,inplace=True)
X.drop('smoothness error',axis=1,inplace=True)
X.drop('concavity error',axis=1,inplace=True)
X.drop('concave points error',axis=1,inplace=True)
X.drop('worst smoothness',axis=1,inplace=True)
X.drop('worst compactness',axis=1,inplace=True)
X.drop('worst fractal dimension',axis=1,inplace=True)
X.drop('mean radius',axis=1,inplace=True)
```

```python
X.shape
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=10,test_size=0.3,shuffle=True)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier(max_features=10)
dt.fit(X_train.values,Y_train)
```

```python
prediction = dt.predict(X_test.values)
```

```python
prediction
```

```python
accuracy_dt=accuracy_score(Y_test,prediction)*100
```

```python
accuracy_dt
```

```python
#knn model
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler().fit(X)
X_std=std.transform(X)
```

```python
sc=StandardScaler().fit(X_train)
X_train_std=sc.transform(X_train)
X_test_std=sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier


knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train_std,Y_train)
```

```python
k_range = range(1,26)
scores = {}

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_std,Y_train)
    prediction_knn = knn.predict(X_test_std)
    scores[k] = accuracy_score(Y_test,prediction_knn)
```

```python
prediction_knn=knn.predict(X_test_std)
```

```python
accuracy_knn=accuracy_score(Y_test,prediction_knn)*100
```

```python
accuracy_knn
```

```python
#Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#model evaluation
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)*100
```

```python
training_data_accuracy
```

```python
#accuracy score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)*100
```

```python
test_data_accuracy
```

```python
#comparison of Algorithm
algorithms=['Decision Tree','KNN','Logistic Regression']
scores=[accuracy_dt,accuracy_knn,test_data_accuracy]


# In[93]:


sns.set(rc={'figure.figsize':(15,7)})
plt.xlabel("Algorithms")
plt.ylabel("Accuracy score")

sns.barplot(data=data_frame,x=algorithms,y=scores)
```
